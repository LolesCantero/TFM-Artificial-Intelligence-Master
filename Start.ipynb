```python
import spacy
```

```python
# Model in spanish
nlp = spacy.load("es_core_news_sm")
```

```python
# Example of text
texto = "Pedro Sánchez visitó Madrid el 15 de marzo de 2024 para asistir a una conferencia de la ONU."
```

```python
# Using spacy with the text
doc = nlp(texto)
```

```python
# tokenization 
print("Tokens:")
for token in doc:
    print(f"{token.text} --> {token.pos_} ({token.dep_})")
```

```python
#lemmatization (base form of each word)
print("\n Lemas:")
for token in doc:
    print(f"{token.text} → {token.lemma_}")
```

```python
# NER
print("\n Entities:")
for ent in doc.ents:
    print(f"{ent.text} → {ent.label_}")
```

```python
#noun chunks
print("\n noun chunks:")
for chunk in doc.noun_chunks:
    print(chunk.text)
```

```python
#vectors
print("\n vecotrs:")
for vec in doc.vector:
    print(vec)
```
