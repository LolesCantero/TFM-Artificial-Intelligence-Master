```python
import spacy
from langdetect import detect
from spacy import displacy  #graph
```

```python
models = {
    "es": spacy.load("es_core_news_sm"),
    "en": spacy.load("en_core_web_sm")
}
```

```python
frases = [
    "Pedro Sánchez visitó Madrid en marzo.",
    "Barack Obama gave a speech in Washington."
]
```

```python
for texto in frases:
    print("=" * 60)
    print(f" Text: {texto}")

    # detect langague
    idioma = detect(texto)
    print(f" Language: {idioma}")

    # Find the model
    if idioma not in models:
        print(" Model not available for this language.")
        continue

    nlp = models[idioma]
    doc = nlp(texto)

    #Tokens
    print("\n Tokens")
    for token in doc:
        print(f"{token.text:15} → POS: {token.pos_:10} | Lemma: {token.lemma_}")

    #NER
    print("\n NER:")
    for ent in doc.ents:
        print(f"{ent.text:25} → {ent.label_}")

    # Phrases
    print("\n Phrases:")
    for chunk in doc.noun_chunks:
        print(f"- {chunk.text}")

    # Graph
    print("\n Graph:")
    displacy.serve(doc, style="dep")
```
